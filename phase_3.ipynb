Advance operation with business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

sale_data = np.array([
    [ 1, 150000, 180000, 220000, 2500001],  # Paradise Biryani
    [ 2, 120000, 20000, 520000, 2900001], # momo
    [ 3, 190000, 180000,920000,2580001], #choco
    [ 4, 100000, 130000,420000,2950001], # pizza
    [ 5, 110000, 80000,2240000,9500001], # burger
    [ 6, 180000, 170000,200000,1500001], #snack     
    ])
print("===sale===",sale_data)
print("\n sales data",sale_data.shape )
print("first three sale",sale_data[0:3])
```

```python
#total sale per year
print(np.sum(sale_data,axis=0))
yearly_total = np.sum(sale_data[:,1:],axis=0)
print(yearly_total)

#minimum sale per year
minimum_sale = np.min(sale_data[:,1:],axis=1)
print(minimum_sale)

#maximim sale per year
max_sale = np.max(sale_data[:,1:],axis=0)
print(max_sale)

#average sale per year
avg_sale = np.mean(sale_data[:,1:],axis=1)
print(avg_sale)
```

```python
##cumulative sum
cumsum = np.cumsum(sale_data[:,1:],axis=1)
print(cumsum)

plt.figure(figsize =(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("average sale ")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show( )
```
