```python
import numpy as np
```

creating array from list


```python
my_list = [1,2,3,4]
print("python list multiplication",my_list * 2)

np_array = np.array([1,2,3,4]) #element wise multiplication
print("python list multiplication",np_array * 2)

import time
start = time.time()
my_list = [i*2 for i in range(100000)]
print("\n list operation time :", time.time() - start)

start = time.time()
np_array = [i*2 for i in range(100000)]
print("\n numpy operation time :", time.time() - start)

```

creating array from scratch

```python
zeros = np.zeros((2,3))
print("\n zeros:",zeros)

ones = np.ones((5,6))
print("\n ones:",ones)

full = np.full((2,5),9)
print("\n full:",full)

random = np.random.random((3,5))
print("\n random:",random)

sequence = np.arange(0,15,2)
print("\n sequence p:",sequence)
```

```python

import numpy as np
arr = np.arange(12)
print("original:",arr)
reshaped = arr.reshape((3,4))
print("\n Resheped",reshaped)

flattened = reshaped.flatten()
print("\n flattened array:",flattened)
#ravel(return views , insted of copy)
raveled = reshaped.ravel()
print("\n raveled array",raveled)

```
